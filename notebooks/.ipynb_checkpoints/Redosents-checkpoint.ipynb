```python
import xml.etree.ElementTree as ET
import os
import re
```

```python
# Rewrite FoLiA XML files
# without the prep attachments
import string

prepdeps = []
dep_on = re.compile(r'<dependencies>')
dep_off = re.compile(r'<\/dependencies>')
depann_on = re.compile(r'<dependency-annotation.*>')
depann_off = re.compile(r'<\/dependency-annotation>')
xmlfiles = []
donedir = '../../data/folia/jblackmore/done'
ids_to_check = []
wrefs_to_check = []
words = []
sents = []
sentids = []
for xmlfile in os.listdir(donedir):
    tree = ET.parse(os.path.join(donedir,xmlfile))
    words = []
    for child in tree.getroot().iter():
        if re.search(r'dependency$',child.tag):
            #print(child.tag)
            depclass = child.attrib['class']
            if 'confidence' in child.attrib:
                #print(child.attrib)
                for attr,val in child.attrib.items():
                    if re.search(r'www.*id$', attr):
                        ids_to_check.append(val)
                        try:
                            wrefs_to_check.append(child[1][0].attrib['id'])
                        except KeyError:
                            print(wrefs_to_check.attrib)
                        
        elif re.search(r'w$',child.tag):
            word_id = None
            for attr,val in child.attrib.items():
                if re.search(r'www.*id$', attr):
                    word_id = val
            if word_id is not None:
                token = child[0]
                words.append(token.text)
        elif re.search(r'folia}s$',child.tag):
            for attr,val in child.attrib.items():
                if re.search(r'www.*id$', attr):
                    if len(words)>0:
                        if words[-1] in string.punctuation:
                            sents.append('{}{}'.format(' '.join(words[:-1]),words[-1]))
                        else:
                            sents.append(' '.join(words))
                    sentids.append(val)
            words = []
    if len(words)>0:
        sents.append('{}{}'.format(' '.join(words[:-1]),words[-1]))
        
```
